Importando os m√≥dulos

```python
from PIL import Image
import pytesseract
import cv2
import numpy as np
```

Carregando a imagem

```python
img = cv2.imread("images/sample2.png")
```

```python
from IPython.display import Image
Image(filename='images/sample2.png')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

Transformacoes para melhorar o reconhecimento dos caracteres

```python
img = cv2.fastNlMeansDenoising(img, None, 3, 11, 33)
img = cv2.resize(img, None, fx=11, fy=11, interpolation = cv2.INTER_CUBIC)
kernel = np.ones((2,2), np.uint8)
img = cv2.erode(img, kernel, iterations=5)
```

```python
from IPython.display import Image
Image(filename='output_image/sample2.png')
```

```python
print( pytesseract.image_to_string(img) ) # Extraindo o texto da imag    em
```
